```python
## Import pandas and matplotlib for data analysis and visualisation
```

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/Lenovo/Downloads/archive (5)/Sample - Superstore.csv", encoding = 'latin-1')
```

```python
data
```

```python
data.describe()
```

```python
data.info()
```

```python
## Changing datatype of columns
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
```

```python
data
```

```python
sales_by_month = round(data.groupby('Order Month')['Sales'].sum(),2)
```

```python
sales_by_month
```

```python
plt.title('Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.plot(sales_by_month, marker = 'o')
plt.show()
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
sns.barplot(x = 'Category', y = 'Sales', data = sales_by_category)
plt.title('Sales by Category')
plt.show()
```

```python
sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_sub_category
```

```python
labels = sales_by_sub_category['Sub-Category']
sales = sales_by_sub_category['Sales']
plt.figure(figsize=(8, 6))
plt.pie(sales, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Sub-Category')
plt.show()
```

```python
profit_by_month = round(data.groupby('Order Month')['Profit'].sum(),2)
```

```python
profit_by_month
```

```python
plt.title('Profit by Month')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.plot(profit_by_month, marker = 'o')
plt.show()
```

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
```

```python
profit_by_category
```

```python
profit_by_sub_category = data.groupby('Sub-Category')['Profit'].sum().reset_index()
```

```python
profit_by_sub_category
```

```python
top_5_profit = profit_by_sub_category.sort_values(by='Profit', ascending=False).head(5)

sns.barplot(x='Sub-Category', y='Profit', data=top_5_profit)
plt.title('Top 5 Sub-Categories by Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=45) 
plt.show()
```
